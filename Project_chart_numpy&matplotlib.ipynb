```python
import numpy as np
from numpy.random import default_rng
import matplotlib.pyplot as plt
from IPython import display

%matplotlib inline
def toy(ball):
    rng = default_rng()
    values = rng.integers(low=1,high=6,endpoint=True,size=(ball))
    return values

while True:
    
    values_chart = np.array([])
    values_dice = list()
    round = 0
    
    print('To stop the program, dial the number 0.')
    round_ball = int(input('How many dice do you want to roll? [balls] : '))
    if round_ball == 0 : break
    round_roll = int(input(f'Want to roll the dice? How many times is {round_ball} balls? [times] : '))
    if round_roll == 0 : break
        
    for round in range(round_roll):
        
        values_dice = toy(round_ball)
        values_chart = np.concatenate((values_chart,values_dice),axis=0)
        plt.hist(values_chart,bins=np.arange(1,8)-0.5,color=('pink'),rwidth=0.9)
        plt.title('Chart Roll Dice',color='brown')
        plt.xlabel('Dice Face (value)',color='brown')
        plt.ylabel('Count (frequency)',color='brown')
        
        if round != (round_roll-1):
            plt.pause(0.1)
        display.display(plt.gcf())
        display.clear_output(wait=True)
    
    plt.show()
```
